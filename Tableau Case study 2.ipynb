# challenge: write a report of all the states with the lowered sales

```python
q = """SELECT 
        State,
        SUM(case when SUBSTRING(([Order Date]), -4, 4) = '2015' then 1 else 0 end) AS [Sum of Sales 2015],
        SUM(case when SUBSTRING(([Order Date]), -4, 4) = '2016' then 1 else 0 end) AS [Sum of Sales 2016],
        SUM(case when SUBSTRING(([Order Date]), -4, 4) = '2017' then 1 else 0 end) AS [Sum of Sales 2017],
        SUM(case when SUBSTRING(([Order Date]), -4, 4) = '2018' then 1 else 0 end) AS [Sum of Sales 2018]
        FROM df
        GROUP BY State
        HAVING "Sum of Sales 2015" < 5
        AND "Sum of Sales 2016" < 5
        AND "Sum of Sales 2017" < 5
        AND "Sum of Sales 2018" < 5
"""
state_orders = pysqldf(q)
state_orders
```
